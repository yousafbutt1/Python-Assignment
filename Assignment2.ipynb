## High_Low_HP

```python
data["HP"].mean()
```

```python
hp_mean = data["HP"].mean()

def set_hp(val):
    if val < hp_mean:
        return "HP Low"
    elif val == hp_mean:
        return "HP neutral"
    else:
        return "HP High"
```

```python
data["HP_high_low"] = data["HP"].apply(set_hp)
data
```

```python
data["Speed"].mean()
```

```python
speed_mean = data["Speed"].mean()

def set_speed(val):
    if val < speed_mean:
        return "Speed Low"
    elif val == speed_mean:
        return "Speed neutral"
    else:
        return "Speed High"
```

```python
data["Speed_high_low"] = data["Speed"].apply(set_speed)
data
```
